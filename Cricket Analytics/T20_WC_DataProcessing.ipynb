# T20 World Cup Cricket Data Pre-Processing #  

```python
#import necessary libraries
import pandas as pd
import json
```

### (1) Process Match Summary ###

```python
with open("t20_wc_match_results.json") as f:
    data = json.load(f)

df_matchSummary = pd.DataFrame(data[0]['matchSummary'])
    
```

```python
df_matchSummary.head(3)
```

```python
df_matchSummary.rename({'scorecard' : 'match_id'}, axis=1, inplace=True)
df_matchSummary.head()
```

```python
match_ids_dict = {}

for index, row in df_matchSummary.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
        
    match_ids_dict[key1] = row["match_id"]
    match_ids_dict[key2] = row["match_id"]
match_ids_dict
```

```python
df_matchSummary.to_csv("matchSummary.csv", index=False)
```

### (2) Process Batting Summary ###


```python
with open("t20_wc_batting_summary.json") as f:
    data = json.load(f)

all_records = []

for rec in data:
    all_records.extend(rec['battingSummary'])

df_battingSummary = pd.DataFrame(all_records)
```

```python
df_battingSummary
```

```python
df_battingSummary['out/not_out'] = df_battingSummary['dismissal'].apply(lambda x : "out" if len(x)>0 else "not_out")
df_battingSummary.drop(columns=['dismissal'], inplace=True)
```

```python
df_battingSummary['batsmanName'] = df_battingSummary['batsmanName'].apply(lambda x : x.replace('â€',''))
df_battingSummary['batsmanName'] = df_battingSummary['batsmanName'].apply(lambda x : x.replace('\xa0',''))
```

```python
df_battingSummary.head()
```

```python
df_battingSummary['match_id'] = df_battingSummary['match'].map(match_ids_dict)
df_battingSummary.head()
```

```python
df_battingSummary.to_csv("battingSummary.csv", index=False)
```

### (3) Process Bowling Summary ###

```python
with open("t20_wc_bowling_summary.json") as f:
    data = json.load(f)

all_brecords = []

for rec in data:
    all_brecords.extend(rec['bowlingSummary'])

df_bowlingSummary = pd.DataFrame(all_brecords)
df_bowlingSummary
```

```python
df_bowlingSummary['match_id'] = df_bowlingSummary['match'].map(match_ids_dict)
df_bowlingSummary.head()
```

```python
df_bowlingSummary.to_csv("bowlingSummary.csv", index=False)
```

### (4) Process Players Information ###


```python
with open("t20_wc_player_info.json") as f:
    data = json.load(f)

df_players = pd.DataFrame(data)
df_players.head(50)
```

```python
df_players['name'] = df_players['name'].apply(lambda x : x.replace('â€',''))
df_players['name'] = df_players['name'].apply(lambda x : x.replace('\xa0',''))
df_players['name'] = df_players['name'].apply(lambda x : x.replace('†',''))
```

```python
df_players.to_csv("playersInformation.csv", index=False)
```
